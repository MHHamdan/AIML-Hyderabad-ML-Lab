### Spectrograms of utterances:

A **spectrogram** is a visual representation of the spectrum of frequencies of a signal as it varies with time. A spectrogram is usually depicted as a heat map, i.e., as an image with the intensity shown by varying the colour or brightness. The objective of this experiment is to observe the spectrograms of utterances of numbers one, two and three. These utterances are provided through .wav files.

###Importing required packages

```python
import librosa
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
from librosa import display
import os
```

###Defining a function to plot spectrogram

```python
def plot_spectrogram(fname, sq, name=None):
    y,sr = librosa.load(fname)
    D = np.abs(librosa.stft(y))
    display.specshow(librosa.power_to_db(D,ref=np.max),y_axis='log',x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    if name is None:
        name = fname.split('/')[-1].split('.')[0]
    plt.title('spectrogram of {}'.format(name))
    plt.tight_layout()
    plt.savefig('spectrogram_of_{}_{}'.format(name, str(int(sq))))
    plt.show()
```

###Plotting the spectrograms i.e., frequency vs time plots for numbers, zero, one and two taken different sample audio files

We can observe that a similar pattern is observed for the set of utterances zeros though they are uttered by different voices. Similarly, for ones and twos utterances. The variation in the position of the pattern depends on the time when the audio is uttered. 

```python
ns = [i for i in os.listdir('./Numbers/') if i != '.DS_Store']
print(ns)
for i in ns:
    DIR = os.path.join('./Numbers', i)
    files = [i for i in os.listdir(DIR) if i != '.DS_Store']
    for index, file in enumerate(files):
        filepath = os.path.join(DIR, file)
        plot_spectrogram(filepath, index, file.split('_')[0])
```
