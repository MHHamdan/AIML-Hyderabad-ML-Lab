## Tossing a Coin n Times

Tossing a coin one or more times is a best example to understand the basics of binomial distribution.


**Binomial Distribution:**

A binomial distribution refers to the probability of getting exact number of successes in a specific number of trials.

* The experiment consists of **n** repeated trials
* Each trial results in only two possible outcomes (Success / Failure)
* The probability of success is denoted by **p**

## Problem Statement

To find out the probability of getting a head when a coin is tossed once or **n** number of times.

**For a coin toss,**

The probability of the outcome Head is $P(Head) = \frac{1}{2} = 0.5$,

Similarly, the probability of the outcome Tail is $P(Tail) = \frac{1}{2} = 0.5$

Hence, **p** = probability of getting head as a success for each toss = $P(Event) = 0.5$

```
# Importing required packages
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
```

### **Tossing a coin once**

Toss a coin and see whether it gives a head or tail. 

Simulate it with the binomial distribution function in Python.

If Number of successes is equal to 1, then it is a head, otherwise it is a Tail.

**np.random.seed()** sets the seed to get the same random number

With the seed reset every time before the np.random operation is performed, the same set of numbers will appear every time.

**Hint:** Refer the [link](https://www.sharpsightlabs.com/blog/numpy-random-seed/) to explore more on the random seed

```
# Set the seed to get the same random number
np.random.seed(42) 

# Simulates tossing a coin once with the probablity of getting head as 0.5
print(np.random.binomial(1,0.5))
```

* *In the above experiment, the outcome of tossing a coin is a tail as the result is zero.*

### **Tossing a coin multiple times**

If a person is tossing a coin 10 times then how many times does the person gets a head in his ten attempts.

```
# Set the seed to get the same random number
np.random.seed(42)

# Tossing a coin 10 times
n = 10

# Probability of success for each toss
p = 0.5

# Simulates tossing a coin 10 times with the probablity of getting head as 0.5
np.random.binomial(n, p)
```

* *In the above experiment, heads occured 4 times in 10 tosses.*

### **Repeating the above coin toss experiment with more number of  persons**

If 100 people are tossing a coin, each 10 times, then how many heads does each person get in his 10 tosses

```
# Set the seed to get the same random number
np.random.seed(42)

# Each person can toss a coin 10 times
n = 10

# Probability of success for each toss
p = 0.5

# 100 People are participating in this event
size = 100 # Replace 100 with the actual number of people present in today's class

# Returns number of successes each person has achieved
total_successes = np.random.binomial(n = n, p = p, size = size)

print("Number of successes (Heads Count) each person achieved:\n", total_successes)
```

### **The Average number of heads in 10 tosses**

A.K.A success rate

```
# Mean of each persons success rate results in the average success rate of the event
print(total_successes.mean())
```
