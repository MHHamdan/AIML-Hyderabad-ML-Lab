# Generate Sentences

Construct a function which receives the list of various types of words as a parameter and generates a sentence.

```
# Declare the list of words with respective weights
determiners = (["The", "A", "Each", "Every", "No"], [0.3, 0.3, 0.1, 0.1, 0.2])
colors = (["red", "green", "blue", "yellow", "grey"], [0.1, 0.3, 0.3, 0.2, 0.2])
nouns = (["water", "elephant", "fish", "light", "programming language"], [0.3, 0.2, 0.1, 0.1, 0.3])
nouns2 = (["of happiness", "of chocolate", "of wisdom", "of challenges", "of air"], [0.5, 0.2, 0.1, 0.1, 0.1])
verb_phrases = (["smells", "dreams", "thinks", "is made of"], [0.4, 0.3, 0.2, 0.1])
```

```
import numpy as np
import math

def generate_sentence(categories):
    words = []
    SPACE = ' '
    
    # From each category select a word and append it to words list
    for word_list, weights in categories:
        
        # Normalize the weights
        normalized_weights = np.array(weights) / sum(weights)
        
        # Numpy random choice helps us pick a random word from the list of word
        selected_word = np.random.choice(word_list, p = normalized_weights)

        words.append(selected_word)
    
    # Join all the words in the list to form a sentence
    return SPACE.join(words)
```

```
# List of all categories of words available
categories = [determiners, colors, nouns, verb_phrases, nouns2]

# Generate 10 sentences
for i in range(10):
    sentence = generate_sentence(categories)
    print(sentence)
```
