Here we will consider a sentence and apply ngram to pick uni-grams, bi-grams, tri-grams and fourgrams.

```
from nltk import ngrams
Sentences=''' One Cent, Two Cents, Old Cent, New Cent: All About Money (Cat in the Hat's Learning Library
        Inside Your Outside: All About the Human Body (Cat in the Hat's Learning Library)
        Oh, The Things You Can Do That Are Good for You: All About Staying Healthy (Cat in the Hat's Learning Library)
        On Beyond Bugs: All About Insects (Cat in the Hat's Learning Library)
        There's No Place Like Space: All About Our Solar System (Cat in the Hat's Learning Library)'''
unigram = list(ngrams(Sentences.lower().split(),6))
bigram=list(ngrams(Sentences.lower().split(),2))
trigram=list(ngrams(Sentences.lower().split(),3))
fourgram=list(ngrams(Sentences.lower().split(),4))
```
