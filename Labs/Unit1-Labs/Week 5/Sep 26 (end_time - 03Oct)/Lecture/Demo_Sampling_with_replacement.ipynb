### Put everything into functions

```
def select_sample(howmany, data):
    return [data[random.randrange(len(data))] for _ in range(howmany)]
```

```
pp.pprint(select_sample(15, dataset))
```

```
def select_samples(nTimes, howmany, data):
    return[select_sample(howmany, data) for _ in range(nTimes)]
```

```
# Generating 5 samples of size 10
pp.pprint(select_samples(5, 10, dataset))
```

```
selections = [sorted(i) for i in select_samples(5, 10, dataset)]
pp.pprint(selections)
```

```
# Joining the selected samples into one string
chosen = ''.join(sorted(''.join([''.join(i) for i in selections])))
print(chosen)
```

```
[chosen.count(ch) for ch in dataset]
```

```
import matplotlib.pyplot as plt
plt.ylabel("Frequency")
plt.xlabel("Letters")
plt.bar(dataset, [chosen.count(ch) for ch in dataset])
plt.show()
```

```
selections2 = [''.join(sorted(i)) for i in select_samples(10, 10, dataset)]
chosen2 = ''.join(selections2)
```

```
plt.ylabel("Frequency")
plt.xlabel("Letters")
plt.bar(dataset, [chosen2.count(ch) for ch in dataset])
plt.show()
```
