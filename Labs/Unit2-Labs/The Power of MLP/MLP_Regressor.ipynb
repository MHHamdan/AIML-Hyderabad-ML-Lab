### Import required packages

```
import pandas as pd
from sklearn.neural_network import MLPRegressor
import numpy as np
```

### PART A:
### Apply MLP regressor for the Linear equation, $y$ = 2 $x_1$ + 3 $x_2$ + 0.5 $x_3$ 



```
x1 = np.random.normal(0, 1, 20)
x2 = np.random.normal(0, 1, 20)
x3 = np.random.normal(0, 1, 20)
y = 2 * x1 + 3 * x2 +  0.5 * x3
```

```
# X is array of x1, x2, x3
X = np.array([x1, x2, x3]).T

# Creating a dataframe with X
linear_data = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
linear_data['y'] = y
linear_data.head()
```

```
# Fit the MLP regressor with Linear Equation data
regr1 = MLPRegressor()
X1 = linear_data.iloc[:,:-1]
Y1 = linear_data.iloc[:, -1]
regr1.fit(X1, Y1)

# Accuracy score
print("Accccuracy is",regr1.score(X1, Y1))
```

### PART B: 

### Apply MLP regressor for the polynomial equation, $y$ = $x_1^3$ + $ 3 x_2^2$ - $ 4 x_3$-2

```
x1 = np.random.normal(0, 1, 20)
x2 = np.random.normal(0, 1, 20)
x3 = np.random.normal(0, 1, 20)

y = x1**3 + 3 * x2**2 - 4 * x3 - 2
```

```
X = np.array([x1, x2, x3]).T
poly_data1 = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
poly_data1['y'] = y
poly_data1.head()
```

```
regr2 = MLPRegressor()
X2 = poly_data1.iloc[:,:-1]
Y2 = poly_data1.iloc[:, -1]
regr2.fit(X2, Y2)

# Accuracy score
print("Accccuracy is",regr2.score(X2, Y2))
```

### PART C:

### Apply MLP regressor for the polynomial equation $y$ = $x_1^4$ - $ 5 x_2^3 $ + $2 x_3 - 7 $

```
x1 = np.random.normal(0, 1, 20)
x2 = np.random.normal(0, 1, 20)
x3 = np.random.normal(0, 1, 20)

y = x1**4 - 5 * x2**3 + 2 * x3 - 7
```

```
X = np.array([x1, x2, x3]).T
poly_data2 = pd.DataFrame(X, columns = ['x1', 'x2', 'x3'])
poly_data2['y'] = y

poly_data2.head()
```

```
regr3 = MLPRegressor()
X3 = poly_data2.iloc[:,:-1]
Y3 = poly_data2.iloc[:, -1]
regr3.fit(X3, Y3)

# Accuracy score
print("Accccuracy is",regr3.score(X3, Y3))
```
