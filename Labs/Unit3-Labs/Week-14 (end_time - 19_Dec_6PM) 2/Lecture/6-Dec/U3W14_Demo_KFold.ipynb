### K-Fold Cross Validation

In k-fold cross-validation, the original sample is randomly partitioned into k equal sized subsamples. Of the k subsamples, a single subsample is retained as the validation data for testing the model, and the remaining k âˆ’ 1 subsamples are used as training data. The cross-validation process is then repeated k times, with each of the k subsamples used exactly once as the validation data. The k results can then be averaged to produce a single estimation.

### Importing required packages

```
#import required packages
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```
x = np.array(["yellow","blue","pink","white","red","violet","orange","green"])
```

```
x
```

```
#Setting our KFold module for 4 splits:
kf = KFold(n_splits=2)
i = 0
# Dividing the data into trainindex and testindex
# K splits are iterated 
for trainindex, testindex in kf.split(x):
    print("Round :",i,":",)
    i += 1
    print("Training index",trainindex,"Training set is:",x[trainindex])
    print("Testing index", testindex," Testing set is:",x[testindex])
```

### **Section II: Train_test_split vs K-fold (Purchase_Dataset)**

```
# Read the data in to a pandas dataframe
data = pd.read_csv('/content/Purchase_Dataset.csv')

print("Shape of data",data.shape)
# Take a look at the first few lines of data to check that we imported correctly
data.head()
```

```
# Converting categorial values to numerical values using label encoder
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```
# X will be our independent variables and y is our dependent/target variable
X = data[['Age','EstimatedSalary','Gender']]
y = data['Purchased']
```

## Splitting the data using train_test_split

```
X_train, X_test, y_train, y_test =  train_test_split(X,y, test_size =0.2, random_state= 2)
print(y_train.values)
print(y_test.values)
```

```
# We instantiate a KNN model, train it and determine the train and test accuracy
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
Testing_accuracy = knn.score(X_test,y_test)
print("Testing accuracy %.2f" %Testing_accuracy)
```

## Splitting the data using Kfold

```
#Setting our KFold module for 5 splits:
kf = KFold(n_splits=10)
```

```
scores_Train = []
scores_Test = []

x_data = X.values
i = 1
for train_index, test_index in kf.split(x_data):
   
    x_train, x_test, y_train, y_test = x_data[train_index], x_data[test_index], y[train_index], y[test_index]
    print("Split :",i,":",)
    i += 1
    print("Distribution of Training labels:",y_train.values)
    print("Distribution of Testing labels:",y_test.values)
    print("--------------------------------------")
    neigh = KNeighborsClassifier()
    neigh.fit(x_train, y_train)
    scores_Train.append(neigh.score(x_train, y_train))
    scores_Test.append(neigh.score(x_test, y_test))
```

```
print("Testing score for each subset %.2f, %.2f, %.2f, %.2f, %.2f,  %.2f, %.2f, %.2f, %.2f, %.2f" %tuple(scores_Test))
#print("Testing score for each subset %.2f, %.2f, %.2f" %tuple(scores_Test))
```

```
#finally, we can average all the scores from all 5 cross validations and get an average score for Train and Test data
print("Average score of the Testing set %.2f"%np.array(scores_Test).mean())
```
