## AI / ML Technique

### Generative Adversary Networks (GAN)


GANs are generative models devised by Goodfellow et al. in 2014. GAN is about creating, like drawing a portrait or composing a symphony. This is hard compared to other deep learning fields. For instance, it is much easier to identify a Monet painting than painting one.


The main focus of GAN is to generate data from scratch, mostly images but other domains including music have been done.

GAN composes of two deep networks :

* Generator
* Discriminator


#### Generator 

The generator tries to produce data that come from some probability distribution. For example, that would be you trying to reproduce the party’s tickets.

#### Discriminator

The discriminator acts like a judge. It gets to decide if the input comes from the generator or from the true training set. For example, that would be the party’s security comparing your fake ticket with the true ticket to find flaws in your design.

In summary, we can say that :

* The generator trying to maximize the probability of making the discriminator mistake its inputs as real.

* And the discriminator guiding the generator to produce more realistic images.

### Importing required  Packages




```
import itertools
import math
import time
import matplotlib.pyplot as plt
from IPython import display

# Pytorch Packages
import torch
import torchvision
import torch.nn as nn
import torchvision.datasets as dsets
import torchvision.transforms as transforms
```

### Loading data

```
transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))])

train_dataset = dsets.MNIST(root='./data/', train=True, download=True, transform=transform)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=100, shuffle=True)
```

### Deffining the Model

```
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            # YOUR CODE HERE : Define a linear layer whose input size is 784 and output size is 1024
            nn.Linear(784, 1024),
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            nn.Dropout(0.3),
            # YOUR CODE HERE : Define a linear layer whose input size is 1024 and output size is 512
            nn.Linear(1024, 512),
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            nn.Dropout(0.6),
            # YOUR CODE HERE : Define a linear layer whose input size is 512 and output size is 256
            nn.Linear(512, 256),
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            nn.Dropout(0.3),
            # YOUR CODE HERE : Define a linear layer whose input size is 256 and output size is 1
            nn.Linear(256, 1),
            # YOUR CODE HERE : Define a 'Sigmoid' activation layer
            nn.Sigmoid()
        )
    
    def forward(self, x):
        out = self.model(x.view(x.size(0), 784))
        out = out.view(out.size(0), -1)
        return out
```

```
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            #YOUR CODE HERE: Define a linear layer whose input size is 100 and output size is 256
            nn.Linear(100, 256),
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            #YOUR CODE HERE: Define a linear layer whose input size is 256 and output size is 512
            nn.Linear(256, 512),
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            #YOUR CODE HERE: Define a linear layer whose input size is 512 and output size is 1024
            nn.Linear(512, 1024), 
            #nn.LeakyReLU(0.2, inplace=True),
            nn.ReLU(),
            #YOUR CODE HERE: Define a linear layer whose input size is 1024 and output size is 784
            nn.Linear(1024, 784),
            #YOUR CODE HERE: Define a 'Tanh' activation layer
            nn.Sigmoid()
            
        )
    
    def forward(self, x):
        x = x.view(x.size(0), 100)
        out = self.model(x)
        return out
```

### Initializing the CUDA

```
use_cuda = torch.cuda.is_available() # YOUR CODE HERE : Check whether CUDA variable is present or not
device = torch.device("cuda" if use_cuda else "cpu")
```

```
discriminator = Discriminator().to(device) # YOUR CODE HERE: Call the model 'Discriminator' defined above and convert it into a CUDA variable.
generator = Generator().to(device) # YOUR CODE HERE: Call the model 'Generator' defined above and convert it into a CUDA variable.
```

### Defining the Optimizer

```
criterion = nn.BCELoss() # YOUR CODE HERE: Initialize BCE Loss function.
lr = 0.0005
d_optimizer = torch.optim.Adam(discriminator.parameters(), lr=lr)
g_optimizer = torch.optim.Adam(generator.parameters(), lr=lr)
```

### Training the Discriminator Function

```
def train_discriminator(discriminator, images, real_labels, fake_images, fake_labels):
    
    # YOUR CODE HERE:  Zero out the gradients using zero_grad() function.
    discriminator.zero_grad() 
    prediction_real = discriminator(images) # YOUR CODE HERE : Call the model object 'discriminator' defined above by passing 'images' as its arguments.
    real_loss = criterion( prediction_real, real_labels)
    prediction_fake = discriminator(fake_images)  # YOUR CODE HERE : Call the model object 'discriminator' defined above by passing 'fake_images' as its arguments. 
    fake_loss = criterion(prediction_fake, fake_labels)
    d_loss = real_loss + fake_loss # Calculate the loss which is sum of 'real_loss' and 'fake_loss'.
    # YOUR CODE HERE : Perform backward pass to compute the gradients
    d_loss.backward()
        
    d_optimizer.step()
    return d_loss,  prediction_real, prediction_fake
```

```
def train_generator(generator, fake_images, real_labels):
    
    # YOUR CODE HERE:  Zero out the gradients using zero_grad() function. 
    generator.zero_grad()
    # Sample noise and generate fake data
    discriminator_outputs = discriminator(fake_images)
    # Calculate error and backpropagate
    g_loss = criterion(discriminator_outputs, real_labels)
    # YOUR CODE HERE : Perform Backward pass
    g_loss.backward()
    g_optimizer.step()
    
    return g_loss
```

### Generate Samples for Testing

```
# Draw samples from the input distribution to inspect the generation on training 
num_test_samples = 16
test_noise = torch.randn(num_test_samples, 100).to(device)
```

### Lets start training, in below output the image gets updated for every iteration

```
# Create figure for plotting
size_figure_grid = int(math.sqrt(num_test_samples))
fig, ax = plt.subplots(size_figure_grid, size_figure_grid, figsize=(6, 6))
for i, j in itertools.product(range(size_figure_grid), range(size_figure_grid)):
    ax[i,j].get_xaxis().set_visible(False)
    ax[i,j].get_yaxis().set_visible(False)

num_epochs = 40 # YOUR CODE HERE : Define number of epochs 
num_batches = len(train_loader) # YOUR CODE HERE : Define number of batches which is equal to the length of train_loader
num_fig = 0 # YOUR CODE HERE : Initialize figure counter


for epoch in range(num_epochs):
    for n, (images, _) in enumerate(train_loader):
        images = images.to(device)
        real_labels = torch.ones(images.size(0), 1).to(device)
        # Sample from generator
        noise = torch.randn(images.size(0), 100).to(device)
        fake_images = generator(noise) # YOUR CODE HERE : Generate a fake image using 'noise' image defined above. HINT : Use 'generator' defined above.
        fake_labels = torch.zeros(images.size(0), 1).to(device)
        # Call 'train_discriminator' define above by passing 'discriminator', 'images', 'real_labels', 'fake_images', 'fake_labels' as its arguments.
        d_loss, d_pred_real, d_pred_fake = train_discriminator(discriminator, images, real_labels, fake_images, fake_labels)
        # Sample again from the generator and get output from discriminator
        noise = torch.randn(images.size(0), 100).to(device)
        # YOUR CODE HERE : Generate a fake image using 'noise' image defined above. HINT : Use 'generator' defined above.
        fake_images = generator(noise)
        # Train the generator
        g_loss = train_generator(generator, fake_images, real_labels)
        
        if (n+1) % 20 == 0:
            test_images = generator(test_noise)
            
            for k in range(num_test_samples):
                i = k//4
                j = k%4
                ax[i,j].cla()
                ax[i,j].imshow(test_images[k,:].data.cpu().numpy().reshape(28, 28), cmap='Greys')
            display.clear_output(wait=True)
            display.display(plt.gcf())
            
            plt.savefig('results/mnist-gan-%03d.png')
            num_fig += 1
            print('Epoch [%d/%d], Step[%d/%d], d_loss: %.4f, g_loss: %.4f, ' 
                  'D(x): %.2f, D(G(z)): %.2f' 
                  %(epoch + 1, num_epochs, n+1, num_batches, d_loss.item(), g_loss.item(),
                    d_pred_real.data.mean(), d_pred_fake.data.mean()))

fig.show()
```
