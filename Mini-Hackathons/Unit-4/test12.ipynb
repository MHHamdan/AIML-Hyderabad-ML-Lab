```
from scipy.special import softmax
import numpy as np
import math 
 
```

```
x_t = np.array([[1, 2, 3]])
x_t_1 = np.array([[1, 2, 3]])
h_t_min_1 = np.array([[0, 0]])

U = np.array([[1,2],
              [2,3],
              [1,1]
                ])

W = np.array([[1,1],
              [2,2]])

V = np.array([[1],
              [5]])
```

```
def fun_h(U,x,W,h):
    r1 = np.matmul(x, U)
    r2 = np.matmul(h, W)
    return r1 + r2
```

```
h_t = fun_h(U,x_t, W, h_t_min_1)
h_t
```

```
o_t = softmax(np.matmul(V, h_t))
o_t
```

```
def softmax(x, axis):
    x -= np.max(x, axis=axis, keepdims=True)
    return np.exp(x) / np.exp(x).sum(axis=axis, keepdims=True)
```

```
x = np.matmul(V, h_t)
softmax(x, axis=1).round(2)
```
